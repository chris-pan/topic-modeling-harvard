```python
import os

dataList = []
with open('/Users/christopherpan 1/Desktop/iscCoeffUnsorted.txt','r') as file1:
    data = file1.readlines()
    for line in data:
        l = line.split(':')
        dataList.append(l)
dataList.sort(key=lambda x: abs(float(x[1])), reverse = True)
```

```python
print(dataList)
```

```python
f = open('/Users/christopherpan 1/Desktop/iscSorted.txt','w')
for data in dataList:
    f.write(str(data[0]) + ':' + str(data[1]))
f.close()
```
