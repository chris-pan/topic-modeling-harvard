```python
import os
rootdir = '/Users/christopherpan 1/Desktop/Topic Modeling Project_SGD/dep/'


cuiDict = []

for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        filename=os.path.join(subdir, file)
        filna = os.path.basename(filename)
        if 'depWords' in filna:
            with open(filename, 'r') as filen:
                data = filen.readlines()
                isFirst = True
                isSecond = True
                for line in data:
                    if isFirst:
                        isFirst = False
                        continue
                    elif isSecond:
                        isSecond = False
                        continue
                    else:
                        l = line.split(',')
                        cui = l[0]
                        if cui in cuiDict:
                            continue
                        else:
                            cuiDict.append(cui)
            print("Done for: "+filna)
```

```python
a['Bin_tag'] = (a['Cos_Sim'] > 0.5)
a['Bin_tag'] = a['Bin_tag'].astype(object).replace({False:'0',True:'1'})
```

```python
clean_frame_train = a.drop(['filename','Cos_Sim'],axis = 1,inplace = False)
clean_frame_train = clean_frame_train.drop(clean_frame_train.columns[0],axis = 1,inplace = False)
```

```python
clean_frame_train
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import normalize

y = clean_frame_train.Bin_tag
#X_train, X_valid, y_train, y_valid = train_test_split(clean_frame_train_2, y, stratify = true, test_size=0.20)
X_train, X_valid, y_train, y_valid = train_test_split(clean_frame_train, y, test_size=0.20)

X_tr = X_train.drop(['Bin_tag'], axis=1)
X_vl = X_valid.drop(['Bin_tag'], axis=1)

X_tr = normalize(X_tr.as_matrix())
X_vl = normalize(X_vl.as_matrix())
#numeric_cols = [col for col in X_tr if X_tr[col].dtype.kind != 'O']
#X_tr[numeric_cols] += 1
#X_tr = np.log(X_tr)
#numeric_cols2 = [col for col in X_vl if X_vl[col].dtype.kind != 'O']
#X_vl[numeric_cols2] += 1
#X_vl = np.log(X_vl)

y_tr = y_train.as_matrix()
y_vl = y_valid.as_matrix()
```

```python
from sklearn import linear_model 
import numpy as np
regr = linear_model.SGDClassifier(loss='log', penalty='elasticnet', alpha=.001, l1_ratio=0.15)
regr.fit(X_tr,y_tr)
a = regr.predict(X_vl)
print(a)
score = regr.score(X_vl, y_vl)
print(score)
```
